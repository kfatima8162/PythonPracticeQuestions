```python
#21. Even or Odd Checker
a=int(input('Enter a number:'))
if(a%2==0):
    print("The given number is even.")
else:
    print("The given number is odd.")
```

```python
#22. Count 4 in List
def count4(list):
    count=0
    
    for i in list:
        if i==4:
            count+=1
    return count

print(count4([4,5,6,2,3,7,1,90,35,67,4,27]))
```

```python
#23. String Prefix Copies
def copy(str,n):
    newstr=str[0:2]
    result=""
    for i in range(n):
        if(len(str)>=2):
            result=result+newstr
        else:
            result=result+str
    return result

print(copy('hello',4))
print(copy('k',4))

        
```
