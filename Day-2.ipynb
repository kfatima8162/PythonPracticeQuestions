```python
#21. Even or Odd Checker
a=int(input('Enter a number:'))
if(a%2==0):
    print("The given number is even.")
else:
    print("The given number is odd.")
```

```python
#22. Count 4 in List
def count4(list):
    count=0
    
    for i in list:
        if i==4:
            count+=1
    return count

print(count4([4,5,6,2,3,7,1,90,35,67,4,27]))
```
