```python
#22. Count 4 in List
def count4(list):
    count=0
    
    for i in list:
        if i==4:
            count+=1
    return count

print(count4([4,5,6,2,3,7,1,90,35,67,4,27]))
```

```python
#23. String Prefix Copies
def copy(str,n):
    newstr=str[0:2]
    result=""
    for i in range(n):
        if(len(str)>=2):
            result=result+newstr
        else:
            result=result+str
    return result

print(copy('hello',4))
print(copy('k',4))

        
```

```python
#25. Value in Group Tester
def test(list,n):
    if n in list :
        return True
    else:
        return False
    
print(test([1,2,3,4,5],6))
print(test(['a','b','c'],'b'))
```
